```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'product':['prod_1', 'prod_2', 'prod_3', 'prod_4', 'prod_5'],
    'division':['A','B','C','B', 'A']
})

df2 = pd.DataFrame({
    'sales':['12500','10800','5600','7900'],
    'product':['prod_3','prod_2','prod_4','prod_1']
})

display(df1, df2)
```

```python
df3 = pd.merge(df1, df2) #detects common column automatically
df3
```

```python
df4 = pd.DataFrame({
    'division':['A','B','C'],
    'manager':['Alex','Rafael','Novak']
})
df4
```

```python
df5 = pd.merge(df3, df4) #one to many
df5
```

```python
df6 = pd.DataFrame({
    'division':['A','A','B','C','C','C'],
    'emp_grade':['13','14+','12','11','10','9-']
})
df6
```

```python
df7 = pd.merge(df1, df6) #many to many
df7 
```

# Merge keys

```python
display(pd.merge(df1, df2, on='product'))
```

```python
df1_new = pd.DataFrame({
    'project':['prod_1', 'prod_2', 'prod_3', 'prod_4', 'prod_5'],
    'division':['A','B','C','B', 'A']
})

display(pd.merge(df1_new, df2, left_on='project', right_on='product'))
display(pd.merge(df1_new, df2, left_on='project', right_on='product').drop('product', axis = 1))
```

```python
df1_index = df1.set_index('product')
df2_index = df2.set_index('product')

display(df1_index, df2_index)
```

```python
display(pd.merge(df1_index, df2_index, left_index=True, right_index=True))
```

```python
df_a = pd.DataFrame({
    'id':['1','2','3','4'],
    'actor_first_name':['rob','gwyn','jon','paul'],
    'actor_last_name':['downey','paltrow','favreau','bettany'],
    'value':['10','6','6','7']
})

df_b = pd.DataFrame({
    'id':['1','2','3','4','5','6'],
    'actor_first_name':['rob','chris','chris','mark','scarlett','jeremy'],
    'actor_last_name':['downey','evans','hemsworth','ruffalo','johanson','renner']
})

display(df_a,df_b)
```

```python
display(pd.merge(df_a,df_b, how='inner'))
display(pd.merge(df_a,df_b, how='left'))
display(pd.merge(df_a,df_b, how='right'))
display(pd.merge(df_a,df_b, how='outer', indicator = True))
```

# Suffixes

```python
df_s1 = pd.DataFrame({
    'product':['prod_1', 'prod_2', 'prod_3', 'prod_4', 'prod_5'],
    'division':['A','B','C','B', 'A']
})

df_s2 = pd.DataFrame({
    'product':['prod_1', 'prod_2', 'prod_3', 'prod_4', 'prod_5'],
    'division':['C','B','B','A', 'B']
})

display(pd.merge(df_s1, df_s2, on='product'))
```

```python
display(pd.merge(df_s1, df_s2, on='product', suffixes=['_ld','_rd']))
```

# Updating dataframes

```python
df1_update = pd.DataFrame({
    'c1':['a','a','b','b'],
    'c2':['x','y','x','y'],
    'val':0
})

df2_update = pd.DataFrame({
    'c1':['a','a','b','b'],
    'c2':['x','y','x','y'],
    'val':[12,31,14,20]
})
```

```python
df1_update.update(df2_update)
df1_update
```

```python
df2_update.combine_first(df1_update)
df2_update
```
