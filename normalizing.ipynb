```python
import pandas as pd
```

```python
wine_data = pd.read_csv('winequality-red.csv', sep=';')
```

```python
wine_data.head()
```

```python
#simple scaling
wine_data['fixed acidity'] = wine_data['fixed acidity'] / wine_data['fixed acidity'].max()
wine_data.head()
```

```python
#min max scaling
#(x - min) / (max - min)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])
wine_data.head()
```

```python
#z-score normalization
#same as standard dcore
#number of standard deviations of a point from the mean (-3 to +3)

```

```python
wine_data['fixed acidity'] = (wine_data['fixed acidity'] - wine_data['fixed acidity'].mean()) / \
    (wine_data['fixed acidity'].std())

wine_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
wine_data[wine_data.columns] = scaler.fit_transform(wine_data[wine_data.columns])

wine_data.head()
```
