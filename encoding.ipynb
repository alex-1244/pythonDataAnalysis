```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
phone_df = pd.DataFrame([
    ['phone 1','blue',2018, 500],
    ['phone 2','black',2016, 450],
    ['phone 3','blue',2017, 600],
    ['phone 4','grey',2018, 800],
])

phone_df.columns = ['Phone','Color','Year','Price']
phone_df
```

```python
from sklearn.preprocessing import LabelEncoder

le_phone = LabelEncoder()
le_color = LabelEncoder()
```

```python
phone_df['phone_enc'] = le_phone.fit_transform(phone_df['Phone'])
phone_df['color_enc'] = le_color.fit_transform(phone_df['Color'])
phone_df
```

```python
from sklearn.preprocessing import OneHotEncoder

oh_phone = OneHotEncoder()
oh_color = OneHotEncoder()

phone_oh = oh_phone.fit_transform(phone_df['phone_enc'].values.reshape(-1,1)).toarray()
color_oh = oh_color.fit_transform(phone_df['color_enc'].values.reshape(-1,1)).toarray()
```

```python
phone_oh
```

```python
color_oh
```

```python
phone_df_one = pd.DataFrame(phone_oh, columns=['Phone_'+str(int(i)) for i in range(phone_oh.shape[1])])
phone_df = pd.concat([phone_df, phone_df_one], axis = 1)
phone_df
```

```python
color_df_one = pd.DataFrame(color_oh, columns=['Color_'+str(int(i)) for i in range(color_oh.shape[1])])
phone_df = pd.concat([phone_df, color_df_one], axis = 1)
phone_df
```

```python
phone_df.drop(['Phone','Color','phone_enc','color_enc'], axis = 1)
```

# Crosstab

```python
adult_df = pd.read_csv('adult.csv')
adult_df.head(10)
```

```python
pd.crosstab(adult_df.occupation, adult_df.education_level)
```

```python
pd.crosstab(adult_df.occupation, adult_df.education_level, margins=True, margins_name='Total')
```

```python
pd.crosstab(adult_df.occupation, adult_df.sex, values=adult_df.age, aggfunc='mean')
```

```python
pd.crosstab([adult_df['marital-status'], adult_df['workclass']], adult_df['sex'])
```

```python
adult_occup_with_total = pd.crosstab(adult_df.occupation, adult_df.education_level, margins=True, margins_name='Total')
display(adult_occup_with_total)
```

```python
adult_occup = adult_occup_with_total.loc[:, adult_occup_with_total.columns != 'Total'].apply(lambda r: r/r.sum(), axis=1)
display(adult_occup)
```

```python
pd.concat([adult_occup, adult_occup_with_total['Total']], axis = 1)
```
