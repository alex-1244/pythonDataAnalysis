```python
survived = by_class['Survived'].sum()
total = by_class['Survived'].count()
survived_vs_count = pd.concat([survived, total], keys=['survived', 'total'], axis = 1)
survived_vs_count
```

```python
survived_vs_count['avg'] = survived_vs_count['survived']/survived_vs_count['total']
survived_vs_count.sort_values(by='avg')
```

# Filter

```python
#mean 'Fare' for group where average age of group > 38 (groupped by Pclass)
by_class.filter(lambda x : x['Age'].mean() > 38)['Fare'].mean()
```

```python
def fillter(x):
    return x['Age'].mean() < 38

by_class.filter(fillter)['Fare'].mean()
```

```python
#mean 'Fare' fro entire datafreme, where average age > 38 (groupped by Pclass)
display(titanic[titanic.Age>38]['Fare'].mean())
display(titanic[titanic.Age<38]['Fare'].mean())
```

```python
titanic['avg_fare_class'] = titanic.groupby('Pclass')['Fare'].transform(lambda x : x.mean())
```

```python
titanic.head()
```

```python
titanic.groupby('Pclass')['Fare']
```

```python
titanic['fare_above_class_avg'] = titanic['Fare'] > titanic['avg_fare_class']
titanic
```

```python
#get count of passengeers who paid above average for their class, and passengers who paid less than avg.
titanic.groupby('fare_above_class_avg').describe().transpose().loc[[('PassengerId','count')]]
```
